```python
import streamlit as st # Handles User interface
import pandas as pd # Builds DF for model
import joblib # Loads saved model

```

```python
model = joblib.load("Final_Forecast_model") # Loads exact model trained
```

```python
st.title('Quantity Demand Forecast App') # Creates title for User Interface

st.write('Enter product + time details to get a prediction') # App instructions 
```

```python
# User Input 


price = st.number_input("Product Price", min_value=0.0, step=0.1)
is_weekend = st.selectbox("Is Weekend?", [0, 1])
month = st.number_input("Month (1-12)", min_value=1, max_value=12, step=1)
year = st.number_input("Year", min_value=2020, max_value=2030, step=1)
day_of_week = st.number_input("Day of Week (0=Mon, 6=Sun)", min_value=0, max_value=6)

prod_qty_mean = st.number_input("Product Avg Quantity", min_value=0.0, step=0.1)
prod_price_mean = st.number_input("Product Avg Price", min_value=0.0, step=0.1)
prod_sales_count = st.number_input("Total Product Sales Count", min_value=0, step=1)

qty_lag1 = st.number_input("Quantity Yesterday (lag1)", min_value=0.0, step=0.1)
qty_lag7 = st.number_input("Quantity Last Week (lag7)", min_value=0.0, step=0.1)

qty_rolling7 = st.number_input("Quantity Rolling 7-day Avg", min_value=0.0, step=0.1)
qty_rolling3 = st.number_input("Quantity Rolling 3-day Avg", min_value=0.0, step=0.1)
```

```python
# Create input dataframe 

input_df = pd.DataFrame([{
    'is_weekend': is_weekend, 
    'day_of_week': day_of_week, 
    'month': month, 
    'year': year, 
    'price': price,
    'prod_sales_count': prod_sales_count, 
    'prod_qty_mean': prod_qty_mean, 
    'prod_price_mean': prod_price_mean,
    'qty_lag1': qty_lag1, 
    'qtylag7': qty_lag7, 
    'qty_rolling3': qty_rolling3,
    'qty_rolling7': qty_rolling7,}])
```

```python
input_df
```

```python
if st.button("Predicted Quantity"):
    prediction = model.predict(input_df)[0]
    st.success(f'Quantity Predicted{prediction: .2f}')

```
